```python
#transform
'''Create a new dataframe containing the best scorers per team (by PTS, total points scored). The resulting dataframe should contain the columns Player, 
Team, Pos and PTS, and should be stored in the variable best_scorers_per_team. It should be sorted by PTS in descending mode.'''
# По заданию мы видим, что нам нужно найти лучшего игрока по очкам в каждой команде. Изначально мы групируем всех играков по названию компанды. Далее, 
# нам нужно каким то образом от каждой команды, найти максимального по количесту очков игрока. Тут мы и будеи использовать transform:
df["Best Score per Team"] = df.groupby('Team')['PTS'].transform('max')

'''Метод transform в библиотеке Pandas используется для применения функции к данным внутри группы и возврата тех же размеров данных, что и 
исходный DataFrame или Series, но с преобразованными значениями.
Когда вы используете transform с методом groupby, он применяет указанную функцию к каждой группе данных независимо и затем возвращает 
результаты обратно в том же формате, что и исходные данные, с сохранением индекса.'''
```

```python
import pandas as pd

# Создаем DataFrame
data = {'Team': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Score': [10, 20, 30, 40, 50, 60]}
df = pd.DataFrame(data)

# Применяем метод groupby и transform для вычисления отклонения от среднего значения
df['mean'] = df.groupby('Team')['Score'].transform('mean')
df['Score_deviation'] = df.groupby('Team')['Score'].transform(lambda x: x - x.mean())

print(df)

'''Этот код вычисляет отклонение каждого значения в столбце 'Score' от среднего значения в соответствующей группе 'Team'. 
Результат сохраняется в новом столбце 'Score_deviation'.'''
```

```python
# это способ получить таблицу, где будут только одно максимальное значение в каждой команде. 
df["Best Score per Team"] = df.groupby('Team')['PTS'].transform('max') #этот код создает колонку с максимальным количеством очков игрока для каждой команды
# очевидно, что одно у нас есть одно значение для каждой компанды. Соответственно , для каждого игрока команды эти значения будут одинаковыми, так как они по дефолту присваиваются.
# но затем мы создаем новую талицу, где уже при помощи сравнения находим того самого игрока: 
best_scorers_per_team = df.loc[df['PTS'] == df["Best Score per Team"], ['Player', 'Team', 'Pos', 'PTS']].sort_values(by = 'PTS', ascending = False)
```

```python
df['Tweet Lower'] = df['Tweet'].str.lower()
В Python, str - это встроенный тип данных, представляющий строки, то есть последовательности символов. 
Когда мы используем str как метод (например, df['Tweet'].str.lower()), мы обращаемся к строковым методам, 
доступным через атрибут str для объектов типа Series в библиотеке Pandas.

Объект типа Series в Pandas представляет собой одномерный массив данных с метками (индексами), и столбцы в 
DataFrame являются объектами типа Series. Когда мы обращаемся к столбцу DataFrame с помощью df['Tweet'], 
мы получаем объект типа Series, который содержит строки.

Таким образом, df['Tweet'].str дает нам доступ к методам строковых операций, которые можно применять к каждой ячейке в 
столбце Tweet. Например, str.lower() преобразует все символы в строке в нижний регистр.
Вот несколько распространенных строковых методов в Python, которые можно использовать с объектами типа Series в библиотеке 
Pandas:

1. `lower()`: Преобразует все символы строки в нижний регистр.
2. `upper()`: Преобразует все символы строки в верхний регистр.
3. `capitalize()`: Преобразует первый символ строки в верхний регистр, а остальные в нижний.
4. `title()`: Преобразует первую букву каждого слова в строке в верхний регистр.
5. `strip()`: Удаляет начальные и конечные пробелы из строки.
6. `lstrip()`: Удаляет начальные пробелы из строки.
7. `rstrip()`: Удаляет конечные пробелы из строки.
8. `replace(old, new)`: Заменяет все вхождения подстроки `old` на подстроку `new`.
9. `split(separator)`: Разбивает строку на список строк по указанному разделителю.
10. `join(iterable)`: Объединяет строки из итерируемого объекта в одну строку, используя текущую строку в качестве разделителя.

Это лишь небольшой список методов. В Python и Pandas существует еще много других методов для работы со строками, 
и иногда их полезно применять в различных ситуациях.
```
